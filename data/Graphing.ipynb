```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# MC

```python
mc_df = pd.read_csv("mc_score_history.csv")

mc_df = mc_df.rename(columns={"Score": "Reward"})
plt.figure(figsize=(8, 5))
plt.plot(mc_df["Move"], mc_df["Reward"], linestyle='-')
plt.xlabel("Move")
plt.ylabel("Reward")
plt.title("Move vs Reward for Monte Carlo Simulation")
plt.grid(True)
plt.show()
```

# DQN

```python
dqn_df = pd.read_csv("dqn_avg.csv")

dqn_df = dqn_df.rename(columns={"Step": "Episode", "Value": "Average Reward"})

plt.figure(figsize=(8, 5))
plt.plot(dqn_df["Episode"], dqn_df["Average Reward"], linestyle='-')
plt.xlabel("Episode")
plt.ylabel("Average Reward")
plt.title("Episode vs Average Reward for DQN Training")
plt.grid(True)
plt.show()
```
